```python
import gymnasium
import matplotlib.pyplot as plt

env = gymnasium.make("ALE/Enduro-v5") #, obs_type="grayscale", full_action_space=True)

obs, _ = env.reset()

plt.imshow(obs, interpolation='nearest')
plt.axis('off')
plt.savefig("media/raw_observation.png", bbox_inches='tight')
obs.shape
```
